```python
import geopandas as gpd
import zipfile
import os

# Path to the zip file and the specific geojson file
zip_path = "input/geoBoundaries-ARG-ADM2-all.zip"
geojson_file = "geoBoundaries-ARG-ADM2.geojson"

# Extract only the specified geojson file to a temporary location
with zipfile.ZipFile(zip_path, "r") as zf:
    zf.extract(geojson_file, "input/tmp")

geojson_path = os.path.join("input/tmp", geojson_file)

# Read with geopandas
gdf = gpd.read_file(geojson_path)
print(gdf.head())
```
